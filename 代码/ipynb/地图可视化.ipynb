```python
#获取站点地理坐标
#data = pd.read_csv('station.csv',use_col=[2,3])
#station = station.drop(index=[31,540]).reset_index()

url = 'http://api.map.baidu.com/geocoder/v2/'
output = 'json'
ak = 'YWdGplhYjUGQ3GtpKNeuTM2S' #你的ak

def getloc(x):
    add = quote(x)
    uri = url + '?' + 'address=' + add  + '&output=' + output + '&ak=' + ak #百度地理编码API
    req = urlopen(uri)
    res = req.read().decode()
    temp = json.loads(res)
    return temp['result']['location']['lng'],temp['result']['location']['lat']#打印出经纬度

a = data['county']
lng = []
lat = []
for i in range(len(a)):
    if(pd.isnull(data['county'][i])== True):
        lng.append(0)
        lat.append(0)
        #print(i)
    else:
        lng.append((getloc(a[i]+'汽车站'))[0])
        lat.append((getloc(a[i]+'汽车站'))[1])
```
