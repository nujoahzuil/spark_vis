```python
import pandas as pd
```

```python
data_g=pd.read_csv("guangdong_station_loc.csv")
data_g.head()
```

```python
data=data_g[["StartCity","ReachCity"]]
```

```python
seattype=data_g['SeatType']
set(seattype)#广东省内全是类型为1的座位
```

```python
from pyecharts.charts import Sankey
from pyecharts import options as opts 
```

```python
num=data.groupby([data_g['StartCity'],data_g['ReachCity']])['ReachCity'].count()
data=data.groupby([data_g['StartCity'],data_g['ReachCity']]).count()
data['num']=num
data.head()
data.to_csv("data_sanky.csv")
```

```python
data_sanky=pd.read_csv("data_sanky.csv")#共20个城市
data_sanky.head(5)
```

```python
for i in range(len(data_sanky['ReachCity'])):
    data_sanky['ReachCity'][i]=data_sanky['ReachCity'][i]+'_'
data_sanky.head(5)
```

```python
nodes = []
for i in set(pd.concat([data_sanky.StartCity, data_sanky.ReachCity])):
    dic_ = {}
    dic_['name'] = i 
    nodes.append(dic_) 
links = []
for x, y, z in zip(data_sanky.StartCity, data_sanky.ReachCity,data_sanky.num):
    dic_ = {}
    dic_['source'] = x
    dic_['target'] = y 
    dic_['value'] = z 
    links.append(dic_) 
```

```python
nodes
```

```python
colors = ['#54B4F9', '#F29150', '#FF7BAE', '#D69AC0', '#485CE0', '#28BE7A'] 
#s = Sankey(init_opts=opts.InitOpts(width='600px', height='500px')) 
s = Sankey()
s.set_colors(colors)
s.add('出发城市-到达城市', 
      nodes, 
      links, 
      linestyle_opt=opts.LineStyleOpts(opacity=0.2, curve=0.5, color='source'), 
      label_opts=opts.LabelOpts(position="left", 
                                font_family='Arial', 
                                margin=10,
                                font_size=13, 
                                font_style='italic')                        
     ) 

s.set_global_opts(title_opts=opts.TitleOpts(title='广东省各市间车流量'))  
s.render('test.html') 
s.render_notebook() 
```
